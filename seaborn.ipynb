```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

#load built in titanic dataset
df = sns.load_dataset('titanic')

print(df.shape)
print(df.head())
print(df.columns.tolist())
```

```python
#1st plot: countplot( count of categorical values)
plt.figure(figsize = (8, 5))
sns.countplot(data = df, x = 'pclass', hue = 'survived')
plt.title("Survival count by passenger class")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.legend(labels=["DIED", "SURVIVED"])
plt.show()
           
```

```python
#histogram with distribution curve
plt.figure(figsize = (8, 5))
sns.histplot(data = df, x = 'age', hue = 'survived', bins = 30, kde = True)
plt.title("Age distribution by survival")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize = (8, 5))
sns.boxplot(data = df, x = 'pclass', y = 'fare', hue = 'survived')
plt.title("Fare distribution by class and survival")
plt.xlabel("Passenger CLass")
plt.ylabel("Fare")
plt.show()
```

```python
#plot 4: heat map
plt.figure(figsize = (8, 5))
correlation = df[['survived', 'pclass', 'age', 'sibsp', 'parch', 'fare']].corr()
sns.heatmap(correlation, annot = True, fmt = '.2f', cmap = 'coolwarm', center = 0)
plt.title('Feature correlation heatmap')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data = df, x = 'age', y = 'fare', hue = 'survived', alpha = 0.6)
plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```
