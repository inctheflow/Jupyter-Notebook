```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

#Generate data
X = np.array([1,2,3,4,5,6,7,8]).reshape(-1,1)
y = np.array([50,55,60,58,61,70,58,90])

x_line = np.linspace(1,8,100).reshape(-1,1)

plt.figure(figsize=(15,4))

#1 under fitting degree = 1
plt.subplot(1,3,1)
model1 = LinearRegression()
model1.fit(X,y)
plt.scatter(X,y)
plt.plot(x_line, model1.predict(x_line))
plt.title("Underfitting (Too simple)")

#2 Good Fit degree = 2
plt.subplot(1,3,2)
poly2 = PolynomialFeatures(degree=2)
X_poly2 = poly2.fit_transform(X)
model2 = LinearRegression()
model2.fit(X_poly2,y)
plt.scatter(X,y)
plt.plot(x_line, model2.predict(poly2.transform(x_line)))
plt.title("Good fit")

#3 Overfitting Curve degree = 6
plt.subplot(1,3,3)
poly6 = PolynomialFeatures(degree=6)
X_poly6 = poly6.fit_transform(X)
model6 = LinearRegression()
model6.fit(X_poly6,y)
plt.scatter(X,y)
plt.plot(x_line, model6.predict(poly6.transform(x_line)))
plt.title("Overfitting curve")

plt.show()
```
