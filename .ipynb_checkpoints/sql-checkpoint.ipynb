```python
import pandas as pd
import sqlite3

#create a database in memory
conn = sqlite3.connect(':memory:')

#create a sample employee table
pd.DataFrame({
    'id': [1, 2, 3, 4, 5, 6],
    'name': ['a', 'b', 'c', 'd', 'e', 'f'],
    'department': ['engineering', 'marketing', 'engineering', 'hr', 'marketing', 'engineering'],
    'salary': [95000, 20000, 99000, 62000, 71000, 102000],
    'age': [28, 20, 55, 29, 25, 52]
}).to_sql('employees', conn, index = False)

print("Database Ready")
```

```python
#select every column when SELECT *
result = pd.read_sql_query("SELECT * FROM employees", conn)
print(result)
```

```python
#select any one column(field)
result = pd.read_sql_query("""
    SELECT name, salary
    FROM employees
""", conn)
print(result)
```

```python
#employee with salary n> 70000
result = pd.read_sql_query("""
    SELECT name, salary
    FROM employees
    WHERE salary > 70000
""", conn)
print(result)
```

```python
#print employees of engineering department in decreasing salary order
result = pd.read_sql_query("""
    SELECT name, salary, department
    FROM employees
    WHERE department == 'engineering'
    ORDER BY salary DESC
""", conn)
print(result)
```

```python
#get name and age of all employees under 30, oder by age
result = pd.read_sql_query("""
    SELECT name, age
    FROM employees
    WHERE age < 30
    ORDER BY age ASC
""", conn)
print(result)
```

```python
#get the name, department and salary of employee who are in marketing OR salary > 90000
result = pd.read_sql_query("""
    SELECT name, department, salary
    FROM employees
    WHERE department = 'marketing' OR salary > 90000
    ORDER BY salary DESC
""", conn)
print(result)
```
