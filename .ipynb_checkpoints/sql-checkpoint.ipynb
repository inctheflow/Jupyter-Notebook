```python
import pandas as pd
import sqlite3

#create a database in memory
conn = sqlite3.connect(':memory:')

#create a sample employee table
pd.DataFrame({
    'id': [1, 2, 3, 4, 5, 6],
    'name': ['a', 'b', 'c', 'd', 'e', 'f'],
    'department': ['engineering', 'marketing', 'engineering', 'hr', 'marketing', 'engineering', 'hr'],
    'salary': [95000, 20000, 99000, 62000, 71000, 102000, 58000],
    'age': [28, 20, 55, 29, 25, 52, 31]
}).to_sql('employees', conn, index = False)

print("Database Ready")
```
