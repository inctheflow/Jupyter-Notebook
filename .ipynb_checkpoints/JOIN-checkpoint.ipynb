```python
import pandas as pd
import sqlite3

conn = sqlite3.connect(':memory:')

#employee table
pd.DataFrame({
    'id': [1, 2, 3, 4, 5, 6],
    'name': ['a', 'b', 'c', 'd', 'e', 'f'],
    'department': ['engineering', 'marketing', 'hr', 'engineering', 'marketing', 'engineering'],
    'salary': [95000, 45000, 50000, 79000, 44000, 90000],
    'age': [45, 22, 27, 29, 19, 31]
}).to_sql('employees', conn, index=False)

#project table
pd.DataFrame({
    'project_id': [1, 2, 3, 4],
    'project_name': ['Website Redesign', 'ML pipeline', 'Mobile APp', 'Data Dashboard'],
    'employee_id': [1, 3, 2, 6],
    'budget': [50000, 120000, 80000, 45000]
}).to_sql('projects', conn, index=False)

print("Both tables ready!")
```

```python
result = pd.read_sql_query("""
    SELECT employees.name, employees.salary, projects.project_name, projects.budget
    FROM employees
    JOIN projects on employees.id = projects.employee_id
""", conn)
print(result)
```

```python
result = pd.read_sql_query("""
    SELECT employees.name, employees.salary, projects.project_name, projects.budget
    FROM employees
    LEFT JOIN projects on employees.id = projects.employee_id
""", conn)
print(result)
```

```python
result = pd.read_sql_query("""
    SELECT employees.name, employees.salary, projects.project_name
    FROM employees
    JOIN projects on employees.id = projects.employee_id
    ORDER BY projects.budget DESC
""", conn)
print(result)
```

```python
result = pd.read_sql_query("""
    SELECT name, salary
    FROM employees
    WHERE salary > (SELECT AVG(salary) FROM employees)
    ORDER BY salary DESC
""", conn)
print(result)
```

```python
#challenge 1 find everyone with same department as person 'a'
result = pd.read_sql_query("""
    SELECT name, id, department
    FROM employees
    WHERE department = (SELECT department FROM employees WHERE name = 'a')
""", conn)
print(result)
```

```python
#challenge 2: find employees whose salary is above average salary of their department
result = pd.read_sql_query("""
    SELECT name, salary, department
    FROM employees
    GROUP BY department
    Having salary > AVG(salary)
""", conn)
print(result)
```

```python
#challenge 3: show each employee's name, salary and difference between their salarya and the company's average
result = pd.read_sql_query("""
    SELECT name, salary, salary - (SELECT AVG(salary) FROM employees) as dif_from_avg
    FROM employees
""", conn)
print(result)


result = pd.read_sql_query("""
    SELECT name, department, salary, 
        AVG(salary) OVER (PARTITION BY department) as dept_avg,
        salary - AVG(salary) OVER (PARTITION BY department) as diff_from_avg
    FROM employees
    ORDER BY department, salary DESC
""", conn)
print(result)
```

```python
result = pd.read_sql_query("""
    SELECT name, department, salary, 
        AVG(salary) OVER (PARTITION BY department) as dept_avg,
        salary - AVG(salary) OVER (PARTITION BY department) as diff_from_avg
    FROM employees
    ORDER BY department, salary DESC
""", conn)
print(result)
```
